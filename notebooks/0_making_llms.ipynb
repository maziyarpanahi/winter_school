```python
from transformers import pipeline

my_pipeline = pipeline(
    "text-classification",
    model="distilbert/distilbert-base-uncased-finetuned-sst-2-english",
)

print(my_pipeline("Wi-Fi is slower than a snail today!"))
```

```python
from transformers import pipeline

# Text Classification (multi-class like news category)
print("\n--- Multi-class Text Classification (News Category) ---")
# Model for news category classification (e.g., AG News dataset)
news_classifier = pipeline(
    "text-classification",
    model="HoYeolY/bert-tiny-finetuned-agnews" # Changed to a publicly accessible model
)
print(news_classifier("The stock market saw a surge in tech shares today."))
print(news_classifier("Lionel Messi scored a stunning goal in the final match."))
```

```python

# Zero-shot Text Classification
print("\n--- Zero-shot Text Classification ---")
# A general zero-shot classification model
zero_shot_classifier = pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli"
)
print(zero_shot_classifier(
    "This is a book about quantum physics.",
    candidate_labels=["finance", "science", "sports", "politics", "art"]
))
```

```python
# Token Classification (NER - Named Entity Recognition)
print("\n--- Token Classification (NER) ---")
# Model for Named Entity Recognition
ner_pipeline = pipeline("ner", model="dslim/bert-base-NER")
print(ner_pipeline("Angela Merkel works in Berlin for the German government."))
```

```python
# Token Classification (POS - Part-of-Speech Tagging)
print("\n--- Token Classification (POS) ---")
# Model for Part-of-Speech Tagging
pos_pipeline = pipeline("token-classification", model="vblagoje/bert-english-uncased-finetuned-pos")
print(pos_pipeline("The quick brown fox jumps over the lazy dog."))
```
